1А

```python
bigram_measures = nltk.collocations.BigramAssocMeasures()
trigram_measures = nltk.collocations.TrigramAssocMeasures()

f = open(r'all.txt', 'r', encoding="UTF-8")
raw = f.read()
f.close()

tokens = nltk.word_tokenize(raw)
#print(tokens[:10])

text = nltk.Text(tokens)
#text.collocations()

def big_trig(text):
    #http://www.nltk.org/_modules/nltk/collocations.html
    finder_bi = BigramCollocationFinder.from_words(text)
    finder_thr = TrigramCollocationFinder.from_words(text)

    # 50 высокочастотных биграмм и триграмм
    big = finder_bi.nbest(bigram_measures.pmi, 50)
    trig = finder_thr.nbest(trigram_measures.pmi, 50)
    return big, trig

big, trig = big_trig(text)

# вывод биграмм и триграмм в файлы
with open('big.csv', 'w') as ouf:
    ouf.write(str(big))
with open('trig.csv', 'w') as ouf:
    ouf.write(str(trig))

# вывод биграмм и триграмм на экран
print(big)
print(trig)
```

Выполнена лемматизация с помощью pymorphy2.

```python
def lemm(token):
    morph = pm.MorphAnalyzer()
    tokens_lem = []
    for token in tokens:
        tokens_lem.append(morph.parse(token)[0].normal_form)
    lems = nltk.Text(tokens_lem)
    return lems
big2, trig2 = big_trig(lems)

# вывод биграмм и триграмм в файлы
with open('big_lem.csv', 'w') as ouf:
    ouf.write(str(big2))
with open('trig_lem.csv', 'w') as ouf:
    ouf.write(str(trig2))

# вывод биграмм и триграмм на экран
print(big2)
print(trig2)
```
