## Задача 1.23. 

Сколько различных маршрутов может избрать пешеход, решивший пройти 9 кварталов, 5 из них - на запад и 4 - на север?

Для решения этой задачи необходимо использовать формулу сочетания с повторениями из n по k, здесь n=9, а k двух типов 4 и 5. Формула сочетаний с повторениями:

$$ {\overline{C}^k}_n = {C^k}_{n+k-1} $$

по условиям задачи $ k = n_1, n_2, ... n_k $

Тогда формула приобретает следующий вид: $$  {{C_n}^{n_1, n_2, ..., n_k}} = \frac{n!}{{n_1}!{n_2}!...{n_k}!} $$

Решение:

$ {C_9}^{4, 5} = \frac{9!}{4!5!} = 126 $

## Задача 2.8

Эксперимент состоит в бросании двух монет. События
<br/> A = { герб на первой монете } ,<br/> B = { цифра на первой монете } ,
<br/> C = { герб на второй монете } ,<br/> D = { цифра на второй монете } ,
<br/> E = { хотя бы один герб } ,<br/> F = { хотя бы одна цифра } ,
<br/> G = { один герб и одна цифра } ,<br/> H = { ни одного герба } ,<br/> K = { два герба } . <br/>
Определить, каким событиям этого списка равносильны следующие события: A + C , AC ,
EF , C + E , CE , BD , E + K , K , CH , FK .

1) Событие А + С означает, что должно произойти хотя бы одно из двух событий, либо оба события. То есть А - выпадет герб на первой монете, или С - выпадет герб на второй монете, или А и С - выпадет два герба.<br\>
Можно увидеть, что должен выпасть хотя бы один герб, что соответствует событию Е.<br\>
А + С = Е

2) Событие АС означает, что должны произойти оба события. То есть А - выпадет герб на первой монете, и С - выпадет герб на второй монете.<br\>
То есть выпадет два герба, что соответствует событию К.<br\>
АС = К

3) Событие ЕF означает, что должны произойти оба события. То есть Е -
выпадет хотя бы один герб, и F - выпадет хотя бы одна цифра.<br\>
Это означает, что выпадет один герб и одна цифра, что соответствует событию G.<br\>
EF = G

4) Событие C + Е означает, что должно произойти хотя бы одно из двух событий, либо оба события. То есть С - выпадет герб на второй монете, или Е - выпадет хотя бы один герб, или С и Е - выпадет хотя бы один герб и он выпадет на второй монете.<br\>
Здесь можно сказать, что событие С включено в событие Е. По аналогии с множествами сложение событий соответствует их объединению. А поскольку событие С является подмножеством множества события Е, то результатом данной суммы будет событие Е.<br\>
С + Е = Е

5) Событие СЕ - означает, что должны произойти оба события. То есть С - выпадет герб на второй монете, и Е - выпадет хотя бы один герб. А поскольку герб должен обязательно выпасть на второй монете, то событие Е уже случилось при условии события С. Таким образом произведение событий С и Е соответствует событию С.
СЕ = С

6) Событие BD - означает, что должны произойти оба события. То есть В - на первой монете выпадет цифра, и D - на второй монете выпадет цифра. Это означает, что выпадут две цифры, то есть ни одного герба, что соответствует событию Н.<br\>
BD = H

7) Событие Е + К - означает, что должно произойти хотя бы одно из двух событий, либо оба события. То есть Е - выпадет хотя бы один герб, К - выпадет два герба. Событие Е означает, что может выпасть как один, так и два герба. Следовательно, сумма событий Е и К соответствует событию Е.<br\>
Е + К = Е

8) Событие $\overline{K}$ означает, отрицание события К. То есть событие К не случится, а значит выпадут не два герба, а хотя бы одна цифра, что соответствует событию F.<br\>
$\overline{K} = F$

9) Событие $\overline{C}H$ означает, что событие С не случится, и вместе с тем произойдет событие Н. То есть $\overline{C}$ - на второй монете не выпадет герб, и Н - не выпадет ни одного герба. Поскольку событие Н обязательно должно произойти и событие $\overline{C}$ не противоречит ему, значит результат будет соответствовать событию Н.<br\>
$\overline{C}H = H$ 

10) Cобытие $\overline{F}K$ означает, что событие F не случится, и вместе с тем проихойдет событие К. То есть $\overline{F}$ - не выпадет хотя бы одна цифра, и К - выпадет два герба. Поскольку событие К пройзойдет и событие $\overline{F}$ не противоречит ему, то результат будет соответствовать событию К.<br\>
$\overline{F}K = K$

## Задача 3.7

Из колоды в 36 карт выбирают 4 карты. Какова вероятность, что 3 из них
красной масти?

Число всех возможных вариантов вытащить 4 карты из 36: ${C_{36}}^4$

Число благоприятных исходов складывается из того, что 3 карты красной масти и 1 карта черной масти. Красных и черных карт в колоде по 18 штук. Поэтому вероятность благоприятных исходов равна: ${C_{18}}^3 {C_{18}}^1$<br\>
Причем ${C_{18}}^1 = 18$

Событие А - выбраны 3 красные и 1 черная карта из колоды 36 карт.<br\>
Вероятность $P(A) = \frac{m}{n}$,<br\>
где m - число благоприятных исходов, n - число всех элементарных исходов.

Решение:<br\>
$P(A) = \frac{{C_{18}}^3 {C_{18}}^1}{{C_{36}}^4} = \frac{14688}{58905} = 0.25 = 25\%$ - вероятность вытащить из колоды 3 красные и 1 черную карту.

## Задача 6.11

В каждой из трех урн содержится 6 черных и 4 белых шара. Из первой
урны наудачу извлечен один шар и переложен во вторую урну, после чего из второй
урны наудачу извлечен один шар и переложен в третью урну. Найти вероятность
того, что шар, наудачу извлеченный из третьей урны, окажется белым.

H1 - из первой урны извлечен белый шар.<br\>
H2 - из первой урны извлечен черный шар.<br\>
$P(H1) = \frac{4}{10} $ - вероятность извлечения белого шара из урны.<br\>
$P(H2) = \frac{6}{10} $ - вероятность извлечения белого шара из урны.<br\>
A - из второй урны извлечен белый шар.<br\>
B - из второй урны извлечен черный шар.<br\>
$P(A|H1) = \frac{5}{11}$ - вероятность того, что из второй урны будет извлечен белый шар, при условии, что из первой урны был извлечен также белый шар.<br\>
$P(A|H2) = \frac{4}{11}$ - вероятность того, что из второй урны будет извлечен белый шар, при условии, что из первой урны был извлечен черный шар.<br\>
$P(B|H1) = \frac{6}{11}$ - вероятность того, что из второй урны будет извлечен черный шар, при условии, что из первой урны был извлечен белый шар.<br\>
$P(B|H2) = \frac{7}{11}$ - вероятность того, что из второй урны будет извлечен черный шар, при условии, что из первой урны был извлечен также черный шар.<br\>

По формуле общей вероятности вычислим вероятность извлечения из второй урны шара того или иного цвета:<br\>
C1 - из второй урны извлечен белый шар.<br\>
С2 - из второй урны извлечен черный шар.<br\>
$P(C1) = P(H1)P(A|H1) + P(H2)P(A|H2) = \frac{4}{10}\frac{5}{11} + \frac{6}{10}\frac{4}{11} = \frac{2}{5}$<br\>
$P(C2) = P(H1)P(B|H1) + P(H2)P(B|H2) = \frac{4}{10}\frac{6}{11} + \frac{6}{10}\frac{7}{11} = \frac{3}{5}$<br\>

$P(D|C1) = \frac{5}{11}$ - вероятность того, что из третьей урны будет извлечен белый шар, при условии, что из второй урны в третью добавлен белый шар.<br\>
$P(D|C2) = \frac{4}{11}$ - вероятность того, что из третьей урны будет извлечен белый шар, при условии, что из второй урны в третью добавлен черный шар.<br\>

Рассчитаем вероятность извлечь белый шар из третьей урны по формуле полной вероятности: <br\>
$P(D) = P(C1)P(D|C1) + P(C2)P(D|C2) = \frac{2}{5}\frac{5}{11} + \frac{3}{5}\frac{4}{11} = \frac{2}{5}$

$P(D) = \frac{2}{5} = 0.4$ - вероятность вынуть из третей урны белый шар.

## Задача 8.9

Среди семян ржи 0, 04% сорняков. Какова вероятность при случайном отборе 5000 семян обнаружить 5 семян сорняков?

Поскольку вероятность найти сорняк р = 0,04% = 0,0004 очень мала, необходимо применять формулу Пуассона: $$P_n(k) \approx \frac{\lambda^k}{k!}e^{-\lambda}$$

Решение:<br\>
n = 5000<br\>
p = 0.0004<br\>
q = 0.9996<br\>
k = 5<br\>
$\lambda = np = 5000 * 0.0004 = 2$

$P_{5000}(5) \approx \frac{2^5}{5!}e^{-2} \approx \frac{32}{120}0.135 \approx 0.036$

## Задача 4.6

Работа прибора прекратилась вследствие выхода из строя одной лампы из общего числа N. Отыскание этой лампы производится путем поочередной замены каждой лампы новой. Определить вероятность того, что придется проверять n ламп, если вероятность выхода из строя каждой лампы равна р.

Поскольку из строя выходит только одна лампа и вероятность выхода ее из строя = р. Все остальные лампы должны быть рабочими. Вероятность, того, что лампа рабочая = 1 - р. При проверке n ламп, необходимо вероятность выхода из строя одной лампы умножить на вероятность не выхода из строя всех остальных n-1 ламп.

Тогда искомая вероятность будет вычисляться по формуле:

$p(1-p)^{n-1}$

## Задача 7.7

Из 18 стрелков 5 попадают в мишень с вероятностью 0.8; 7 - с вероятностью 0.7; 4 - с вероятностью 0.6; и 2 - с вероятностью 0.5. Наудачу выбранный стрелок произвел выстрел, но в мишень не попал. К какой из групп вероятнее всего принадлежал этот стрелок?

Гипотеза H1 - стрелял стрелок из первой группы<br\>
$P(H1) = \frac{5}{18}$<br\>
Гипотеза H2 - стрелял стрелок из второй группы<br\>
$P(H2) = \frac{7}{18}$<br\>
Гипотеза H3 - стрелял стрелок из третьей группы<br\>
$P(H3) = \frac{4}{18}$<br\>
Гипотеза H4 - стрелял стрелок из четвертой группы<br\>
$P(H4) = \frac{2}{18}$<br\>
Событие A - стрелок не попал.

$P(A|H1) = 1 - 0.8 = 0.2$ - вероятность того, что стрелок из первой группы промахнется.<br\>
$P(A|H2) = 1 - 0.7 = 0.3$ - вероятность того, что стрелок из второй группы промахнется.<br\>
$P(A|H3) = 1 - 0.6 = 0.4$ - вероятность того, что стрелок из третьей группы промахнется.<br\>
$P(A|H1) = 1 - 0.5 = 0.5$ - вероятность того, что стрелок из четвертой группы промахнется.<br\>
<br\>
Вычислим вероятность для каждой гипотезы по формуле Байеса:<br\>
$P(H_i|A) = \frac{P(H_i)P(A|H_i)}{P(A)}$, где $P(A) = \sum{P(H_i)P(A|H_i)}$<br\>
<br\>
$P(A) = \frac{5}{18}\frac{2}{10}+\frac{7}{18}\frac{3}{10}+\frac{4}{18}\frac{4}{10}+\frac{2}{18}\frac{5}{10} = \frac{57}{180}$<br\>
<br\>
$P(H1|A) = \frac{P(H1)P(A|H1)}{P(A)} = \frac{\frac{5}{18}\frac{2}{10}}{\frac{57}{180}} = \frac{10}{57}$ - вероятность промаха стрелка из первой группы.<br\>
$P(H2|A) = \frac{P(H2)P(A|H2)}{P(A)} = \frac{\frac{7}{18}\frac{3}{10}}{\frac{57}{180}} = \frac{21}{57}$ - вероятность промаха стрелка из второй группы.<br\>
$P(H3|A) = \frac{P(H3)P(A|H3)}{P(A)} = \frac{\frac{4}{18}\frac{4}{10}}{\frac{57}{180}} = \frac{16}{57}$ - вероятность промаха стрелка из третьей группы.<br\>
$P(H4|A) = \frac{P(H4)P(A|H4)}{P(A)} = \frac{\frac{2}{18}\frac{5}{10}}{\frac{57}{180}} = \frac{10}{57}$ - вероятность промаха стрелка из четвертой группы.<br\>

Ответ: вероятнее всего стрелял стрелок из второй группы.
