### Задание 3. Обработка отсутствующих значений.

```python
dataset = pd.read_csv('travel-times.csv', sep=',')
dataset.head(10)
```

```python
dataset.info()
```

```python
dataset['FuelEconomy'] = dataset['FuelEconomy'].apply(lambda x: np.NaN if x == '-' else np.float64(x))
dataset['FuelEconomy'] = dataset['FuelEconomy'].fillna(dataset['FuelEconomy'].mean())
```

```python
dataset.head(10)
```

```python
dataset.to_csv('dataset.csv')
```
